1. Use lambda with filter to select numbers greater than 5 from [3, 6, 9, 2, 5].

```python
# Write answer here:

arr=[3, 6, 9, 2, 5]
a=filter(lambda x:x>5,arr)

print(list(a))
```

2. Use map and lambda to capitalize all strings in ['apple', 'banana', 'cherry'].

```python
# Write answer here:
arr=['apple', 'banana', 'cherry']
a=map(lambda a:a.capitalize(),arr)

print(list(a))
```

3. Use zip and lambda to calculate element-wise average of [10,20,30] and [20,30,40].

```python
# Write answer here:
a=[10,20,30]
b=[20,30,40]

c=map(lambda x:(x[0]+x[1])/2,zip(a,b))

print(list(c))
```

4. Use map and lambda to flatten a list of lists [[1,2],[3,4],[5,6]].

```python
# Write answer here:

nested = [[1, 2], [3, 4], [5, 6]]

print ([item for sublist in nested for item in sublist])

flattened = list(map(lambda x: x, [item for sublist in nested for item in sublist]))
print(flattened)
```

5. Use filter and lambda to keep strings containing the letter 'e' in ['apple','banana','cherry'].

```python
# Write answer here:

arr=['apple', 'banana', 'cherry']

a=filter(lambda x:"e" in x,arr)

print(list(a))
```

6. Use zip and lambda to subtract corresponding elements of [10,20,30] and [1,2,3].

```python
# Write answer here:

a=[10, 20, 30]
b=[1,2,3]

a=map(lambda x:x[0]-x[1],zip(a,b))

print(list(a))
```

7. Use map, filter, and lambda together to square only even numbers from [1,2,3,4,5,6].

```python
# Write answer here:

arr=[1,2,3,4,5,6]

a=map(lambda x:x**2,filter(lambda x:x%2==0,arr))

print(list(a))
```

8. Use map and lambda to extract the first character of each string in ['apple','banana','cherry'].

```python
# Write answer here:

arr=["apple", "banana", "cherry"]

a=map(lambda x:x[1:],arr)

print(list(a))
```

9. Use filter and lambda to remove empty strings from ['', 'apple', '', 'banana'].

```python
# Write answer here:

arr=['', 'apple', '', 'banana']

a=filter(lambda x:len(x)!=0,arr)

print(list(a))
```

10. Use zip, map, and lambda to multiply three lists element-wise [1,2], [3,4], [5,6].

```python
# Write answer here:
a=[1,2]
b=[3,4]
c=[5,6]

a=map(lambda x:x[0]*x[1]*x[2],zip(a,b,c))

print(list(a))
```
