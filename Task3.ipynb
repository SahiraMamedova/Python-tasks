1. Write a program to check if a number is positive, negative, or zero.

```python
# Write answer here:

def checkNum(num):
    if num>0:
        print("positive")
    elif num==0:
        print("zero")
    else:
        print("negative")

checkNum(1)
```

2. Write a program to check if a number is multiple of 3 and 7.

```python
# Write answer here:

def isMultiple(num):
    if num%3==0 and num%7==0:
        print("yes")
    else: print("no")

isMultiple(21)
isMultiple(8)
```

```python
# Write answer here:

def checkDigit(num):
    if(len(str(num))>1):
        print("multiple")
    else: print("single")

checkDigit(31)
checkDigit(3)
```

7. Write a program to find the middle number of three numbers.

```python
# Write answer here:

#this solution in not with sort

def findMiddle(num1,num2,num3):
    if (num1>num2 and num1<num3) or (num1>num3 and num1<num2):
        print("Middle is",num1)
    elif (num2>num1 and num2<num3) or (num2>num3 and num2<num1):
        print("Middle is",num2)
    else: print("Middle is",num3)

findMiddle(1,7,2)
```

8. Write a program to check if a given point (x, y) lies inside, outside, or on a circle of radius r.

```python
# Write answer here:


def checkPoint(x,y,r):
    if x**2+y**2==r**2:
        print("on the circle")
    elif x**2+y**2<r**2:
        print("inside circle")
    else: print("outside circle")

checkPoint(2,3,4)
```

9. Write a program to assign letter grades with plus/minus, and add “honors” if average > 90.

```python
# Write answer here:

average=93

if average > 97:
    grade = "A+"
elif average > 93:
    grade = "A"
elif average > 90:
    grade = "A-"
elif average > 87:
    grade = "B+"
elif average > 83:
    grade = "B"
elif average > 80:
    grade = "B-"
elif average > 77:
    grade = "C+"
elif average > 73:
    grade = "C"
elif average > 70:
    grade = "C-"
elif average > 60:
    grade = "D"
else:
    grade = "F"

# Add honors
if average > 90:
    grade += " (Honors)"

print(grade)
```
