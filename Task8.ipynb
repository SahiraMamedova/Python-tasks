Mean

```python
arr=[1,100,4,2,34,21,33]
```

```python
def mean (data):
    sum=0
    for i in range(len(data)):
        sum+=data[i]
    return sum/len(data)
```

```python
print(mean(arr))
```

Median

```python
def median(data):
    length=len(data)
    midindex=len(data)//2
    data.sort()
    if length%2==0:
        return (data[midindex]+data[midindex-1])/2
    else:
        return data[midindex]
```

```python
print(median(arr))
```

Mode

```python
def mode(data):
    dict={}
    maxCount=0
    for i in range(len(data)):
        strNum=str(data[i])
        if strNum not in dict:
            dict[strNum]=0
        else:
            dict[strNum]+=1
    arr=sorted(dict.values())
    for i in range(len(arr)):
        if arr[i]>maxCount:
            maxCount=arr[i]
    valArr=[key for key,value in dict.items() if value==maxCount and maxCount!=0]
    return "No mode" if len(valArr)==0 else valArr
```

```python
print(mode(arr))
```
