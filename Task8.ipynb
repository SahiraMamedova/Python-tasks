Mean

```python
arr=[1,100,4,2,34,21,33]
```

```python
def mean (data):
    sum=0
    for i in range(len(data)):
        sum+=data[i]
    return sum/len(data)
```

```python
print(mean(arr))
```

Median

```python
def median(data):
    length=len(data)
    midindex=len(data)//2
    data.sort()
    if length%2==0:
        return (data[midindex]+data[midindex-1])/2
    else:
        return data[midindex]
```

```python
print(median(arr))
```

Mode

```python
def mode(data):
    dict={}
    maxCount=0
    for i in range(len(data)):
        strNum=str(data[i])
        if strNum not in dict:
            dict[strNum]=0
        else:
            dict[strNum]+=1
    arr=sorted(dict.values())
    for i in range(len(arr)):
        if arr[i]>maxCount:
            maxCount=arr[i]
    valArr=[key for key,value in dict.items() if value==maxCount and maxCount!=0]
    return "No mode" if len(valArr)==0 else valArr
```

```python
print(mode(arr))
```

Q1 Q2 AND Q3

```python
def findQuartiles(arr):
    sortedarr = sorted(arr)
    n = len(sortedarr)

    # positions (1-based)
    pos_q1 = (n + 1) / 4
    pos_q2 = (n + 1) / 2
    pos_q3 = 3 * (n + 1) / 4

    def get_value(pos):
        if pos.is_integer():
            return sortedarr[int(pos) - 1]  
        else:
            lower = sortedarr[int(pos) - 1]
            upper = sortedarr[int(pos)]
            return (lower + upper) / 2

    q1 = get_value(pos_q1)
    q2 = get_value(pos_q2)
    q3 = get_value(pos_q3)

    return q1, q2, q3
  
q1, q2, q3 = findQuartiles(arr)
print(f"Q1: {q1}, Q2: {q2}, Q3: {q3}")
```

Standart deviation is square root of variance.

```python
def findVariance(arr):
    m=mean(arr)
    sum=0
    for i in range(0,len(arr)):
        sum+=(arr[i]-m)**2
    return sum/len(arr)

variance=findVariance(arr)
print("Sample variance",variance)

print("Standart deviation",variance**0.5)
```
