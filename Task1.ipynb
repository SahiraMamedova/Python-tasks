7. Reverse a string using a loop (no slicing).

```python
# Write answer here:

str="hello"

reversedStr=""

for i in range(len(str)-1,-1,-1):
    reversedStr+=str[i]

print(reversedStr)
```

8. Find factorial of a number using a loop.

```python
# Write answer here:

num=6

fact=1

for i in range (1,num+1):
     fact*=i

print(fact)
```

9. Find the largest number in a list using a loop.

```python
# Write answer here:

largestNum=0

listArr=[34,12,90,23,56]

for i in range(0,len(listArr)):
    if(listArr[i]>largestNum):
        largestNum=listArr[i]

print(largestNum)
```
