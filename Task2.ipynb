2. Write a function to return both sum and average of a list.

```python
# Write answer here:

listArr=[1,2,3,4,5]

sum=0

sumFn=lambda beforeVal,val: beforeVal+val

for i in range(0,len(listArr)):
    sum=sumFn(sum,listArr[i])

print("Sum",sum,"Average",sum/len(listArr))
```

3. Write a function to find the maximum number in a list.

```python
# Write answer here:

listArr=[1,2,3,4,5]

largestNum=0

findLargest=lambda num1,num2:num1>num2

for i in range(0,len(listArr)):
    if(findLargest(listArr[i],largestNum)):
        largestNum=listArr[i]

print(largestNum)
```

4. Write a function to reverse a string.

```python
# Write answer here:


def reverseStr(str):
    reversedStr=""
    for i in range(len(str)-1,-1,-1):
        reversedStr+=str[i]
    return reversedStr
    
print(reverseStr("helllo"))
```

5. Write a function to check if a number is prime or not.

```python
# Write answer here:

primeOrEeven = lambda x: "Even" if x % 2 == 0 else "Prime"

print(primeOrEeven(2))
```

6. Write a function to generate the first n Fibonacci numbers.

```python
# Write answer here:

def fibonacci(n):
    a, b = 0, 1 
    for i in range(n):
        print(a)  
        a, b = b, a + b  

fibonacci(10)
```

7. Write a function to calculate the area of a circle given the radius.

8. Write a function to count the number of words in a sentence.

```python
# Write answer here:


example="Hi , I am sahira"

def countWords(sentence):
    punctuation = ",.!?;:'\""

    for i in punctuation:
        sentence=sentence.replace(i,"")
        
    return len(sentence.split())

print(countWords(example))
 
```

9. Write a function to return a list of even numbers from a given list.

```python
# Write answer here:

listArr=[1,5,2,7,8]

isEven=lambda x:x%2==0

result=filter(isEven,listArr)

print(list(result))
```

```python
# Write answer here:

# fromkeys creates dictionary from list making list element as key, and default value as None

removeDuplicates=lambda listExample:list(dict.fromkeys(listExample))

print(removeDuplicates([1,4,2,6,7,3,4,2,4]))
```
