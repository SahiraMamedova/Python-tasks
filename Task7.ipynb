125.check palindrome

```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        i=0
        cleanedstr=""
        while i<len(s):
            if s[i].isalnum():
                cleanedstr+=s[i]
            i+=1
        return "".join(reversed(cleanedstr)).lower()==cleanedstr.lower()       

        
```

392.first occurence

```python
class Solution:
    def isSubsequence(self, s: str, t: str) -> bool:
        for i in range(len(s)):
            if s[i] not in t:
                return False
        return True
```

28.first occurence

```python
class Solution:
    def strStr(self, haystack: str, needle: str) -> int:
        for i in range(len(haystack)):
            if haystack[i:len(needle)]==needle:
                return i
        return -1
```

169.majority element

```python
class Solution:
    def majorityElement(self, nums: List[int]) -> int:
        dict={}
        for i in range(len(nums)):
            num=str(nums[i])
            if num not in dict:
                dict[num]=1
            else:
                dict[num]+=1
            if dict[num]>len(nums)/2:
                return nums[i]
        
```

58. Length of Last Word

```python
class Solution:
    def lengthOfLastWord(self, s: str) -> int:
        trimmedstr=s.strip()
        splitted=trimmedstr.split(" ")
        last=splitted[len(splitted)-1]
        return len(last)
```
